```python
#RGB to gray scale
im_gray = im.convert('L')
im_hist_gray = im_gray.histogram()
plt.hist(im_hist_gray)
plt.show()
```

```python
#Valor RGB para un pixel especÃ­fico
im.getpixel((0,0))
```

```python
#mezclo los canales r,g,b -> b,g,r
r, g, b = im.split()
#im_invert = Image.merge("RGB", (b, g, r))
#im_invert.show()
```

```python
plt.close()
im_r_hist = r.histogram()
im_g_hist = g.histogram()
im_b_hist = b.histogram()
im_rgb_hist = im.histogram()
plt.hist(im_r_hist)
plt.hist(im_g_hist)
plt.hist(im_b_hist)
#plt.hist((im_r_hist, im_g_hist, im_b_hist)) #bars side by side no stacked
plt.show()
```

```python
plt.close()


fig = plt.figure(figsize=(10,5))

fig.suptitle('Histograma RGB', fontsize=18, y=1.05, horizontalalignment='center')

ax1 = fig.add_subplot(331) #fila columna orden
ax1.plot(im_r_hist, color='red')
plt.title('Red Channel')
ax2 = fig.add_subplot(332)
ax2.plot(im_g_hist, color='green')
plt.title('Green Channel')
ax3 = fig.add_subplot(333)
ax3.plot(im_b_hist, color='blue')
plt.title('Blue Channel')

gs = gridspec.GridSpec(3, 3)

ax4 = fig.add_subplot(gs[1,:3])
ax4.plot(im_rgb_hist, color='black')
plt.title('RGB')

ax5 = fig.add_subplot(gs[2,:3])
ax5.plot(im_hist_gray, color='gray')
plt.title('Gray Scale')

fig.tight_layout()

#plt.subplot(211)
#plt.plot(im_r_hist, im_g_hist)
#plt.plot(im_r_hist, color = 'red')
#plt.plot(im_g_hist, color = 'green')
#plt.plot(im_b_hist, color = 'blue')
#plt.plot(im_hist_gray, color = 'gray')
plt.show()
```

```python
plt.close()
fig, ax = plt.subplots(4, figsize=(6, 12))
ax[0].hist(im_r_hist, color = 'red')
ax[1].hist(im_g_hist, color = 'green')
ax[2].hist(im_b_hist, color = 'blue')
ax[3].hist(im_hist_gray, color = 'gray')
#fig.show()
fig.suptitle('Histograma RGB', fontsize=18, y=1.03)
fig.tight_layout()
fig.savefig("prueba.png")
plt.show()
```
